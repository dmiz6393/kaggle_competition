Estimators - These are almost exaclty synonymous to how we use the word models in language. <br>
Transformers - Are robots in disguise
    - They alter/transform our features

Transformers follow this type code pattern:
```python
# instantiate the object
transformer = TransformerClass()

# fit estimator on X_train and transform X_train
transformer.fit_transform(X_train)

# transform X_test
transformer.fit(X_test)
```

Estimators follow _this_ code pattern:
```python
# instantiate the object
estimator = EstimatorClass()

# fit based on training data
estimator.fit(X_train, y_train)

# getting our predictions
estimator.predict(X_train)
```

````python
def r2_adj(X,y): 
    lr= LinearRegression()
    model=lr.fit(X,y)
    r_squared=model.score(X,y)
    adjusted_r2_formula= 1 - (1 - r_squared)*(len(y)-1)/(len(y)-X.shape[1]-1)
    print(adjusted_r2_formula)
````

````python
r2_adj(X,y)
````
