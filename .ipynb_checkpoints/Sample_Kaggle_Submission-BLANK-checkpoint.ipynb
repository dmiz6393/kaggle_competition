## Preprocessing and Modeling

```python
# Use "Overall Qual", "Lot Area", "Street" as features
#by looking at the correlations, potential factors: 
# YEAR BUILT-.57
# MAS VNR AREA - .51
# total bsmnt sf- .63 
# 1st floor SF- .62 
# gr liv area- .7
# full bath- .5  
# TotRms AbvGrd-.50 
# Garage Yr Blt .53 
# garage cars .65
# garage area- .65
# OVERALL QUAL-.8
# YEAR REMOD/ADD - .55 

#First, I'd like to create a simple model with two features with the highest correlations: overall qual and gr liv area
features = ["Overall Qual","Gr Liv Area"]
X = df_train[features]
y = df_train["SalePrice"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=9)
```

```python
# Convert the street categories into processable numbers (i.e. turn categories into dummy variables)
lr=LinearRegression()
```

```python
# Train-Test Split
cross_val_score(lr,X_train,y_train,cv=5).mean()
```

```python
lr.fit(X_train,y_train)
```

```python
predictions=lr.predict(X_test)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
